## Ingesta de datos
La Base de Datos es games.csv, la cual contiene informacion referida a juegos de ajedrez, en ella vemos los ID de los jugadores, cantidad de movimientos, ganador, etc.

```python
# Importamos las librerias necesarias
import pandas as pd

# Leemos el CSV y luego pasamos a mostrarlo
df = pd.read_csv('./parte_2_games.csv')
```

```python
# Mostramos las 5 primeras filas
df.head()
```

```python
# Mostramos las 5 ultimas filas
df.tail(10)
```

```python
# Mostramos la dimension
df.shape
```

```python
# Mostramos los tipos de datos de las columnas
df.dtypes
```

```python
# Vemos si hay algun dato missings
df.count()
```

```python
# Dejaremos solo las columnas que no sean object
columnas = ['rated', 'created_at', 'last_move_at', 'turns', 'white_rating', 'black_rating', 'opening_ply']
df_filtrado = df[columnas]
df_filtrado.head()
```

```python
# Listado de las 10 partidas con mas movimientos
df_filtrado.nlargest(10,'turns')
```

```python
# Mostramos las cantidades de los distintos tipos de finales
df['victory_status'].value_counts().plot(kind = 'bar')
```

```python
# Cuales son los 10 "primeros movimientos" de blacos y negros
df['opening_name'].value_counts().head(10).plot(kind = 'bar')
```
